```python
list_df = []
df = pd.DataFrame(columns=['Stock', 'Date','Open','High','Low','Close','Volume','Adj Close','Close_norm'])
```

```python
entries = 0
for name in files:
    df_temp = pd.read_csv('dataset/' + name + '.csv')
    df_temp.loc[:,'Stock'] = name
    min = df_temp.loc[:,'Close'].min()
    max = df_temp.loc[:,'Close'].max()
    df_temp.loc[:,'Close_norm'] = (df_temp.loc[:,'Close']-min)/(max-min)
    list_df.append(df_temp)
    entries += df_temp.shape[0]
    #df = df.append(df_temp, ignore_index=True)
    
bip()
entries
```

```python
df = pd.concat(list_df)
```

## Correlations

```python
df[df.loc[:,'Stock']=='AAL'].describe()
```

```python
def fn(group):
    corr = list()
    for i in range(1,91):
        corr.append(group.loc[:,'Close_norm'].autocorr(i))
    autocorr_df.loc[group.name,1:90] = corr
```

```python
autocorr_df = pd.DataFrame(columns=range(1,91), index=files)
df.groupby('Stock').apply(fn);
bip()
```

```python
df[df.loc[:,'Stock'] == 'A']
```

```python
autocorr_df.loc['A']
```

```python
autocorr_df.loc[:,'var'] = autocorr_df.var(axis=1)
```

```python
autocorr_df.T.plot(figsize=(15,7))
```

# 10 Highest Variance Stocks

```python
plot = autocorr_df.sort_values('var', ascending=False)[:10].T.plot(figsize=(15,7), title="10 highest variance stocks")
plot.set_xlabel("# of lags")
plot.set_ylabel("autocorr")
```

# 10 lowest variance stocks

```python
plot = autocorr_df.sort_values('var', ascending=True)[:10].T.plot(figsize=(15,7), title='10 lowest variance stocks')
plot.set_xlabel("# of lags")
plot.set_ylabel("autocorr")
```

## Highest autorrelation values

```python
def get_max(row):
    max = row.max()
    max_name = row.idxmax()
    row_sorted = row.sort_values(ascending=False)
    #bp()
    autocorr_df.loc[row.name,'max_autocorr_val'] = max
    autocorr_df.loc[row.name,'1st_highest'] = max_name
    autocorr_df.loc[row.name,'2nd_highest'] = row_sorted.index[1]
    autocorr_df.loc[row.name,'3th_highest'] = row_sorted.index[2]
    autocorr_df.loc[row.name,'4th_highest'] = row_sorted.index[3]
    autocorr_df.loc[row.name,'5th_highest'] = row_sorted.index[4]
```

```python
autocorr_df.iloc[:,0:90].apply(get_max, axis=1);
bip()
```

## Scatter plot of highest autocorrelations

```python
plt.figure(figsize=(20,7))
plot = plt.scatter(x=autocorr_df.loc[:,'1st_highest'],y=autocorr_df.loc[:,'max_autocorr_val'], alpha=0.1, s=200)
plt.title('Scatter plot of the highest correlations and its lags')
plt.ylabel('Autocorr')
plt.xlabel('# of lags')
```

## Histograms

```python
plt.figure(figsize=(20,5))
plt.hist(x=[
    autocorr_df.loc[:,'1st_highest'],
    autocorr_df.loc[:,'2nd_highest'],
    autocorr_df.loc[:,'3th_highest'],
    autocorr_df.loc[:,'4th_highest'],
    autocorr_df.loc[:,'5th_highest'],
], stacked=True, label=[1,2,3,4,5],rwidth=0.1, align='left');
plt.title('Histogram of 5th largest autocorrelations')
plt.xticks(range(0,91,2))
plt.legend()
```
